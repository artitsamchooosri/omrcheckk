```python
import numpy as np
import cv2
from IPython import display
from PIL import Image
import matplotlib.pyplot as py
import ipywidgets as widgets
%matplotlib inline

def clicked(frame,img_counter):
    img_name = "opencv_frame_{}.png".format(img_counter)
    cv2.imwrite(img_name, frame)
    print("{} written!".format(img_name))



def showVideo(VideoIndex=0, scale=0.5):
    try:
        cap = cv2.VideoCapture(VideoIndex)
    except:
        print("Cannot Open Device")
    try:
        ret, frame = cap.read()
        
        while(ret==True):
            ret, frame = cap.read()
            
            if not ret:
                cap.release()
                print ("Released Video Resource")
                break
            half_frame = cv2.resize(frame, (0, 0), fx = scale, fy = scale)
            #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            _,ret_array = cv2.imencode('.jpg', half_frame) 
            i = display.Image(data=ret_array)
            display.display(i)
            button_download = widgets.Button(description = 'Test Button')   
            button_download.on_click(clicked(frame,1))
            display.display(button_download)
            display.clear_output(wait=True)
    except KeyboardInterrupt:
        cap.release()
        print("Released Video Resource from KeyboardInterrupt")
    pass
```

```python
showVideo()
```
