```python
import cv2
import numpy as np
import fnc
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io
import os
from imutils.perspective import four_point_transform
```

```python
def splitBoxes(img,_row,_col):
    rows = np.vsplit(img, _row)
    boxes = []
    for r in rows:
        cols = np.hsplit(r, _col)
        for box in cols:
            boxes.append(box)
    return boxes
def splitBoxes_t(img):
    rows = np.vsplit(img, 1)
    boxes = []
    for r in rows:
        cols = np.hsplit(r, 5)
        for box in cols:
            boxes.append(box)
    return boxes
def crop_aws(img):
    green = (255, 0, 0)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    blur = cv2.GaussianBlur(gray, (5, 5), cv2.BORDER_DEFAULT)
    edged = cv2.Canny(blur, 100, 200)
    contours, _ = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE )
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    cv2.drawContours(img, contours, -1, green, 1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    edged = cv2.Canny(gray, 350, 50)
    contours, _ = cv2.findContours(edged, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE )
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    for contour in contours:
        peri = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, 0.05 * peri, True)
        if len(approx) == 4:
            doc_cnts = approx
            break
    warped = four_point_transform(img, doc_cnts.reshape(4, 2))
    warped = cv2.resize(warped, (1280, 1280))
    y = 20
    x = 20
    h = 1260
    w = 1260
    crop = warped[y:y + h, x:x + w]
    warped = cv2.resize(crop, (1260, 1260))
    return warped
```

```python
old_working_dir = os.getcwd()
pathImage=old_working_dir+'\\img'
partchoice=old_working_dir+"\\choice"

#i=0
fname = []
for (dirpath, dirnames, filenames) in os.walk(pathImage):
    fname.extend(filenames)
    break
print(fname)
for f in fname:
    print(f)
    if '.jpg' in f:
        p_img=pathImage+"\\"+f
        img = io.imread(p_img)
        img = cv2.resize(img, (700, 700))
        img_aws=crop_aws(img)
            #plt.imshow(img_aws)
        Boxes_Row=[]
        Boxes_Col= splitBoxes(img_aws,1,4)
        for r, Row in enumerate(Boxes_Col):
            Boxes_Row.extend(splitBoxes(Boxes_Col[r],21,1))
        Boxes_Row.pop(0)
        Boxes_Row.pop(20)
        Boxes_Row.pop(40)
        Boxes_Row.pop(60)
        for r, Row in enumerate(Boxes_Row):
            status = mpimg.imsave(partchoice+"\\"+f+"_"+str(r)+".jpg",Boxes_Row[r])
            print("Image OK written to file-system : ",status)
```
